## Amazon Web Scraper with Email Automation

This Amazon Scraper allows you to scrape the price of Dre Beats Headphones on Amazon. It also allows you to email the headphone listing to any email address. With this project I was able to scrape the data of Name, Price, Description of the Item. I also used the automation of sending an email if the pricing is below $300.00. The email address "BlackDevsRock@Gmail.Com" two-step authorization was turned on and random password was generated using (Google App Password). 


```python
import requests
```

```python
from bs4 import BeautifulSoup
```

```python
import smtplib
```

```python
URL = 'https://www.amazon.com/Beats-Solo3-Wireless-Ear-Headphones/dp/B01LWWY3E2/ref=sr_1_7?crid=1LIFVO252KQJS&dchild=1&keywords=beats+by+dre+wireless+headphones&qid=1598576180&sprefix=Beats+by+dre%2Caps%2C193&sr=8-7'
```

```python
headers = {"User-Agent": 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/84.0.4147.135 Safari/537.36'}
```

```python
page = requests.get(URL, headers=headers)
```

```python
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
print(soup.prettify)
```

```python
def check_price():
    title = soup.find(id="productTitle").get_text()
    price = soup.find(id="priceblock_ourprice").get_text()
    description = soup.find(id="feature-bullets").get_text()
    converted_price = float(price[1:7])
    
    if(converted_price < 350.00):
        send_mail()
        
    print(converted_price)
    print(title.strip())
    print(description.strip())
        
    if(converted_price > 350.00):
        send_mail()
    
def send_mail():
        server = smtplib.SMTP('smtp.gmail.com', 587)
        server.ehlo()
        server.starttls()
        server.ehlo()
        
        server.login('BlackDevsRock@gmail.com', 'tcpaxtfpienagjds')
        
        subject = 'Price Fell Down'
        body = 'Check the Amazon link:https://www.amazon.com/Beats-Solo3-Wireless-Ear-Headphones/dp/B01LWWY3E2/ref=sr_1_7?crid=1LIFVO252KQJS&dchild=1&keywords=beats+by+dre+wireless+headphones&qid=1598576180&sprefix=Beats+by+dre%2Caps%2C193&sr=8-7'
        msg = f"Subject: {subject}\n\n{body}"
        
        server.sendmail(
            'blackdevsrock@gmail.com',
            'designisorion@gmail.com',
            msg
        )
        print('New Headphone Listing. Enjoy!')
        
        server.quit()
    
check_price()
        
    
```
